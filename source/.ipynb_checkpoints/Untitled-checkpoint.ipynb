```python
import pandas as pd
from numpy.random import randint
```

```python
TablaDummies = pd.DataFrame({'IDDummie':['D1','D2','D3','D4','D5','D6'],
                             'Edad':['28', '53', '59', '18', '18', '18'],
                             'Género':['M', 'F', 'M', 'F', 'F', 'F']                             })
TablaGov = pd.DataFrame({'ID':['112020685','1513131','15646'],
                         'Edad':['28', '53','18'],
                         'Género':['M', 'F','F'],
                         'C19':['1', '1','1']})


TablaDummies = TablaDummies.sort_values('Edad',ascending=False)
TablaGov = TablaGov.sort_values('Edad',ascending=True)

TablaDummies['Key'] = TablaDummies['Edad'] + TablaDummies['Género']
TablaGov['Key'] = TablaGov['Edad'] + TablaGov['Género']
```

```python
TablaGov
```

```python
TablaDummies
```

```python
T = pd.merge(TablaDummies,
              TablaGov[['Key','C19', 'ID']],
                 on='Key', 
                 how='left')

T
```

```python
T = T.sort_values('ID',ascending=True)
T
```

```python
T = T.drop_duplicates(subset=['ID'], keep='first')

T
```

```python
del TablaDummies['Key']

TablaDummies
```

```python
TM1 = pd.merge(TablaDummies,
              T[['C19', 'IDDummie']],
                 on='IDDummie', 
                 how='left')

TM1
```

```python
TM1 = TM1.fillna(0)

TM1
```
